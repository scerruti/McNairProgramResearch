# Phase‑2 Control Analysis (cerruti)

This notebook reproduces the Sprint‑2 heatmap comparison (e.g. `height` vs `rotation`) and implements a random‑group control: we randomly split the *same set of questions* into two groups (matching the original category sizes) and compute the same magnitude metric for the difference between their aggregated expert activation vectors.

The goal is to show whether the observed category difference is larger than what we would expect by random partitioning of the same questions.


```python
# Configuration
RESULT_JSON = 'data/phase2/runpod_second/results.json'  # fallback location for Sprint-2 results
HEATMAP_DIR = 'data/phase2/runpod_second'
CATEGORY_A = 'height'
CATEGORY_B = 'rotation'
N_PERMUTATIONS = 2000  # default; increase if you want higher precision
OUTPUT_DIR = 'data/phase2/runpod_second/analysis_controls'
import os
os.makedirs(OUTPUT_DIR, exist_ok=True)
print('Output dir:', OUTPUT_DIR)
```

```python
# Common imports
import json
from collections import Counter, defaultdict
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

# Helper: load results.json (per-question expert frequencies and categories)
def load_results(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f'Results JSON not found at {path}')
    with open(path, 'r') as f:
        report = json.load(f)
    per_q = report.get('per_question_results') or report.get('per_question') or {}
    return per_q

per_question = load_results(RESULT_JSON)
print('Loaded per-question entries:', len(per_question))
```

```python
# Build question lists and aggregated expert vectors for the two categories
questions_by_category = defaultdict(list)
for qid, q in per_question.items():
    cat = q.get('category')
    if cat is None:
        continue
    questions_by_category[cat].append(qid)

print({k: len(v) for k,v in questions_by_category.items()})
if CATEGORY_A not in questions_by_category or CATEGORY_B not in questions_by_category:
    raise ValueError('One of the chosen categories not found in results JSON')

qids_A = questions_by_category[CATEGORY_A]
qids_B = questions_by_category[CATEGORY_B]
nA, nB = len(qids_A), len(qids_B)
print(f'Category sizes: {CATEGORY_A}={nA}, {CATEGORY_B}={nB}')

# helper to aggregate expert counts across a list of qids
def aggregate_counts(qid_list):
    agg = Counter()
    for qid in qid_list:
        q = per_question.get(str(qid)) or per_question.get(qid) or {}
        freq = q.get('expert_frequency') or q.get('image_expert_frequency') or {}
        for eid_str, cnt in freq.items():
            try:
                eid = int(eid_str)
            except Exception:
                eid = eid_str
            agg[eid] += int(cnt)
    return agg

agg_A = aggregate_counts(qids_A)
agg_B = aggregate_counts(qids_B)
# unified expert index list
expert_ids = sorted(set(list(agg_A.keys()) + list(agg_B.keys())))
vec_A = np.array([agg_A.get(e,0) for e in expert_ids], dtype=float)
vec_B = np.array([agg_B.get(e,0) for e in expert_ids], dtype=float)

print('Total activations (A):', vec_A.sum(), ' (B):', vec_B.sum())
```

```python
# Compute observed difference metric between categories
diff = vec_A - vec_B
fro_norm = np.linalg.norm(diff)
mean_abs = np.mean(np.abs(diff))
max_abs = np.max(np.abs(diff))
print('Observed metrics: Frobenius norm =', fro_norm)
print(' mean abs diff =', mean_abs, ' max abs diff =', max_abs)

# Plot difference bar (top contributors)
top_idx = np.argsort(np.abs(diff))[::-1][:30]
top_experts = [expert_ids[i] for i in top_idx]
top_vals = diff[top_idx]
plt.figure(figsize=(12,6))
plt.bar([str(e) for e in top_experts], top_vals, color=['tab:blue' if v>0 else 'tab:orange' for v in top_vals])
plt.axhline(0, color='k', linewidth=0.6)
plt.title(f'Top 30 experts by difference ({CATEGORY_A} - {CATEGORY_B})')
plt.xlabel('expert id')
plt.ylabel('activation difference')
plt.xticks(rotation=45)
plt.tight_layout()
out_png = os.path.join(OUTPUT_DIR, f'top30_diff_{CATEGORY_A}_vs_{CATEGORY_B}.png')
plt.savefig(out_png, dpi=200, bbox_inches='tight')
print('Saved', out_png)
plt.show()
```

```python
# Permutation control: randomly split the combined set of questions from the two categories
import random
combined_q = list(qids_A) + list(qids_B)
N = N_PERMUTATIONS
metrics = []
rng = random.Random(12345)  # deterministic seed for reproducibility
for t in range(N):
    rng.shuffle(combined_q)
    randA = combined_q[:nA]
    randB = combined_q[nA:nA+nB]
    agg_randA = aggregate_counts(randA)
    agg_randB = aggregate_counts(randB)
    vec_randA = np.array([agg_randA.get(e,0) for e in expert_ids], dtype=float)
    vec_randB = np.array([agg_randB.get(e,0) for e in expert_ids], dtype=float)
    diff_rand = vec_randA - vec_randB
    metrics.append(np.linalg.norm(diff_rand))
    if (t+1) % 200 == 0:
        print('Done', t+1)

metrics = np.array(metrics)
pval = (np.sum(metrics >= fro_norm) + 1) / (len(metrics) + 1)
print('Permutation p-value (>= observed fro_norm):', pval)

# Save metrics and histogram
import pandas as pd
df_metrics = pd.DataFrame({'fro_norm': metrics})
df_metrics.to_csv(os.path.join(OUTPUT_DIR, 'permutation_metrics.csv'), index=False)

plt.figure(figsize=(8,4))
plt.hist(metrics, bins=50, alpha=0.8)
   
```

```python
plt.axvline(fro_norm, color='red', linewidth=2, label='observed')
plt.xlabel('Frobenius norm of difference (random split)')
plt.ylabel('count')
plt.title('Permutation distribution of difference metric (random splits)')
plt.legend()
hist_png = os.path.join(OUTPUT_DIR, f'permutation_hist_{CATEGORY_A}_vs_{CATEGORY_B}.png')
plt.tight_layout()
plt.savefig(hist_png, dpi=200, bbox_inches='tight')
print('Saved', hist_png)
plt.show()
```

Interpretation notes:

- If the observed Frobenius norm is well within the permutation distribution, the category difference magnitude is consistent with random partitioning of the same questions (i.e., no meaningful category signal).
- If the observed norm is in the far tail (small p-value), the category assignment corresponds to a larger difference than random groups, supporting non-random structure.

You can increase `N_PERMUTATIONS` for a more precise p-value, or change the metric (mean absolute difference, KL, or top-K overlap) if you prefer a different sensitivity.
